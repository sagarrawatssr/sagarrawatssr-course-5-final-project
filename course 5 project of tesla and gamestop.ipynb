#  Reset the index, save, and display the first five rows of the tesla_data dataframe using the head function 

```python
import matplotlib.dates
import pandas as pd
import yfinance as yf
ts=yf.Ticker("TSLA")
tesla_data1=ts.history(period="max")
tesla_data=tesla_data1.reset_index()
tesla_data.head()
```

# displaying the last 5 rows of the revenue of tesla using tail function

```python
tesla_all=ts.financials
tesla_all=tesla_all.transpose()
tesla_all.reset_index(inplace=True)
tesla_all=tesla_all.rename(columns={"index":"Date"})
tesla_all["Date"]=pd.to_datetime(tesla_all['Date']).dt.tz_localize('UTC')
tesla_revenue=tesla_all.loc[:,"Total Revenue"]
tesla_revenue.tail()
```

# Reset the index, save, and display the first five rows of the gme_data dataframe using the head function

```python
gs=yf.Ticker("GME")
gs_data=gs.history(period="max")
gme_data=gs_data.reset_index()
gme_data.head()
```

# displaying the last 5 rows of the revenue of gamestop using tail function

```python
gme_all=gs.financials
gme_all=gme_all.transpose()
gme_all.reset_index(inplace=True)
gme_all=gme_all.rename(columns={"index":"Date"})
gme_all["date"]=pd.to_datetime(gme_all["Date"]).dt.tz_localize('UTC')
gme_revenue=gme_all.loc[:,"Total Revenue"]
gme_revenue.tail()
```

# plot the graph of tesladata and revenue

```python
import matplotlib.pyplot as plt

def make_graph(ts_close,ts_revenue,ts_stdate,ts_redate):
    ## to set the figure of whole plot here we will set the width and height of whole graph.
    f = plt.figure()
    f.set_figwidth(12.4)
    f.set_figheight(5)

    # to plot the graph in 1 row and 2 column where stock price graph is at 1 position
    plt.subplot(1,2,1)
    plt.plot(ts_stdate,ts_close,"-r",linewidth=1,label="stock price")
    plt.xlabel("date",color='b')
    plt.ylabel("stock price",color='b')
    plt.title("graph of tesla stock")
    
    # to set the legend of label given in plot attribute
    
    plt.legend()
    # to plot the graph in 1 row and 2 column where stock revenue graph is at 2 position
   
    plt.subplot(1,2,2)
    
    plt.plot(ts_redate,ts_revenue,"-b",linewidth=2,label="stock revenue")
   # plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    plt.gcf().autofmt_xdate()

    plt.xlabel("date",color='b')
    plt.ylabel("revenue",color='b')
    plt.title("graph of tesla revenue")

    # given the legends here also
    plt.legend()

    # show the graph
    
    plt.show()

# extract whole required data here to pass to the make_graph function

''' here is predefined code for revenue
    tesla_all=ts.financials
    tesla_all=tesla_all.transpose()
    tesla_all.reset_index(inplace=True)
    tesla_all=tesla_all.rename(columns={"index":"Date"}) '''
   # tesla_revenue=tesla_all.loc[:,"Total Revenue"]

ts_close=tesla_data.Close
ts_stdate=tesla_data.Date
ts_redate=tesla_all.Date

#calling the graph to execute

make_graph(ts_close,tesla_revenue,ts_stdate,ts_redate)
```
